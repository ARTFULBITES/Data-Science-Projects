```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 2, 3, 4],
    'B': [5, 6, 7, 8, 9],
    'C': [9, 5, 3, 3, 2]
}
df = pd.DataFrame(data)

# Number of rows and columns for subplots
n_cols = len(df.columns)
n_rows = 1  # Change this if you have many columns

# Create figure and axes objects
fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(5 * n_cols, 5))

# Plot each column
for i, col in enumerate(df.columns):
    if n_cols > 1:
        ax = axs[i]  # Use axs[i] if multiple columns
    else:
        ax = axs  # Use axs directly if only one column
    
    ax.hist(df[col], bins=5, color='blue', alpha=0.7)
    ax.set_title(f'Histogram of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

# Improve layout to prevent overlap
plt.tight_layout()
plt.show()
```
